#📌 **Extracción**

### Importando archivos JSON

```python
import pandas as pd
import requests

url = 'https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/main/TelecomX_Data.json'

response = requests.get(url)
data_json = response.json()
df = pd.DataFrame(data_json)

print("Datos cargados exitosamente:")
df.head()
```

### 🔍 Explicación del Dataset

<div style="background-color: #f8f9fa; padding: 20px; border-left: 6px solid #1565c0; border-radius: 5px; font-family: sans-serif;">
  
  <p>Al ejecutar el comando <code>df.head()</code>, se observa que el DataFrame cuenta únicamente con <b>6 columnas</b>. Sin embargo, de acuerdo con el diccionario de datos, el conjunto debería integrar muchas más variables (como <i>gender, MonthlyCharges, Contract</i>, etc.).</p>

  <p style="background-color: #fff3cd; padding: 10px; border-radius: 4px; border: 1px solid #ffeeba;">
     <b>Presunción:</b> Los datos en el archivo JSON se encuentran en una estructura <b>anidada</b>.
  </p>

  <p>Se confirma visualmente que la información relevante está encapsulada dentro de diccionarios en las siguientes columnas:</p>
  
  <ul>
    <li><code>customer</code></li>
    <li><code>phone</code></li>
    <li><code>internet</code></li>

</div>

### Normalizando archivos JSON

```python
customer_df = pd.json_normalize(df['customer'])
phone_df = pd.json_normalize(df['phone'])
internet_df = pd.json_normalize(df['internet'])
account_df = pd.json_normalize(df['account'])

# 2. Concatenamos los resultados con las columnas base (customerID y Churn)
df_normalizado = pd.concat([
    df[['customerID', 'Churn']],
    customer_df,
    phone_df,
    internet_df,
    account_df
], axis=1)

df_normalizado.head()
```

## Manejo de inconsistencias

### Paso 1: Limpieza de Nulos y espacios en blanco

```python
df_normalizado['Churn'] = df_normalizado['Churn'].str.strip()
df_limpio = df_normalizado[df_normalizado['Churn'] != ""].copy()
df_limpio = df_limpio.dropna(subset=['Churn', 'Charges.Total'])
```

### Paso 2: Estandarización de Inconsistencias Categóricas

```python
# En este dataset, 'No internet service' es redundante si ya tenemos la columna InternetService
# Lo estandarizamos a 'No' para simplificar el análisis
cols_con_redundancia = [
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

for col in cols_con_redundancia:
    df_limpio[col] = df_limpio[col].replace({'No internet service': 'No'})
```

### Paso 3: Verificación de Resultados

```python
print(f"Registros tras limpieza: {len(df_limpio)}")
print(f"Nulos en Churn: {df_limpio['Churn'].isnull().sum()}")
print(f"Tipo de dato de Churn: {df_limpio['Churn'].dtype}")
```

```python
df_limpio.head()
```

## Columna de cuentas diarias (Opcional)

### 1. Calculamos la facturación diaria dividiendo los cargos mensuales entre 30

```python
df_limpio['Cuentas_Diarias'] = (df_limpio['Charges.Monthly'] / 30).round(2)
```

### 2. Verificación de la nueva métrica

## Estandarización y transformación de datos (opcional)

### 1. Diccionario de renombrado para mayor claridad (Stakeholders)

```python
columnas_espanol = {
    'customerID': 'ID_Cliente',
    'gender': 'Genero',
    'SeniorCitizen': 'Adulto_Mayor',
    'Partner': 'Tiene_Pareja',
    'Dependents': 'Personas_a_Cargo',
    'tenure': 'Meses_Antiguedad',
    'PhoneService': 'Servicio_Telefonico',
    'MultipleLines': 'Lineas_Multiples',
    'InternetService': 'Servicio_Internet',
    'OnlineSecurity': 'Seguridad_Online',
    'OnlineBackup': 'Respaldo_Online',
    'DeviceProtection': 'Proteccion_Dispositivo',
    'TechSupport': 'Soporte_Tecnico',
    'StreamingTV': 'Streaming_TV',
    'StreamingMovies': 'Streaming_Peliculas',
    'Contract': 'Tipo_Contrato',
    'PaperlessBilling': 'Facturacion_Electronica',
    'PaymentMethod': 'Metodo_Pago',
    'Charges.Monthly': 'Cargos_Mensuales',
    'Charges.Total': 'Cargos_Totales',
    'Churn': 'Evasion'
}

df_limpio = df_limpio.rename(columns=columnas_espanol)
```

### 2. Mapeo de variables binarias (Sí/No -> 1/0)

```python
cols_binarias = [
    'Tiene_Pareja', 'Personas_a_Cargo', 'Servicio_Telefonico',
    'Facturacion_Electronica', 'Evasion'
]

for col in cols_binarias:
    df_limpio[col] = df_limpio[col].map({'Yes': 1, 'No': 0}).astype(int)
```

### 3. Traducción de categorías dentro de las columnas para mejor interpretación

```python
mapeo_contratos = {
    'Month-to-month': 'Mensual',
    'One year': 'Anual',
    'Two year': 'Bianual'
}
df_limpio['Tipo_Contrato'] = df_limpio['Tipo_Contrato'].replace(mapeo_contratos)
```

### 4. Verificación final

```python
df_limpio.head()
```

## Distribución de evasión

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
total = len(df_transformado)
counts = df_transformado['Evasion'].value_counts()
percentages = (counts / total * 100)

plt.figure(figsize=(8, 6))
sns.set_style("white")
ax = sns.barplot(x=percentages.index, y=percentages.values, hue=percentages.index, palette=['#45B39D', '#EC7063'], legend=False)

plt.title('Distribución de Clientes por Estado de Evasión', fontsize=16, pad=20)
plt.ylabel('Porcentaje (%)', fontsize=12)
plt.xlabel('')

# Reemplazar 0/1 por etiquetas descriptivas en el eje X
ax.set_xticklabels(['Clientes Activos (0)', 'Clientes Perdidos (1)'], fontsize=12)

plt.ylim(0, 110)

# Añadir etiquetas de porcentaje sobre las barras
for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}%',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 10),
                textcoords='offset points',
                fontsize=12,
                fontweight='bold')

sns.despine()
plt.show()
```

## Recuento de evasión por variables categóricas

```python
variables_analizar = ['Genero','Tipo_Contrato', 'Metodo_Pago', 'Servicio_Internet']

plt.figure(figsize=(20, 8))
sns.set_style("white")

for i, var in enumerate(variables_analizar):
    plt.subplot(2, 2, i+1)

    # 0: Clientes Activos, 1: Clientes Perdidos
    ax = sns.countplot(x=var, hue='Evasion', data=df_transformado, palette=['#45B39D', '#EC7063'])

    plt.title(f'Evasión por {var.replace("_", " ")}', fontsize=14, fontweight='bold')
    plt.xlabel('', fontsize = 11)
    plt.ylabel('Cantidad de Clientes')
    plt.ylim(0, 3000)

    # Ajustar leyenda
    if i == 0:
        plt.legend(title='Estado', labels=['Activo', 'Perdido'])
    else:
        ax.get_legend().remove()

    # Añadir etiquetas de valor sobre las barras
    for p in ax.patches:
        height = p.get_height()
        if height > 0:
            ax.annotate(f'{int(height)}',
                        (p.get_x() + p.get_width() / 2., height),
                        ha='center', va='center',
                        xytext=(0, 7),
                        textcoords='offset points',
                        fontsize=10)

sns.despine()
plt.tight_layout()
plt.show()
```

### Matriz de Correlación

```python
import numpy as np
import matplotlib.colors as mcolors

cols_corr = [
    'Evasion', 'Meses_Antiguedad', 'Cargos_Mensuales', 'Cargos_Totales',
    'Cuentas_Diarias', 'Adulto_Mayor', 'Tiene_Pareja', 'Personas_a_Cargo',
    'Servicio_Telefonico', 'Facturacion_Electronica'
]

cmap_personalizado = mcolors.LinearSegmentedColormap.from_list(
    "propio", ["#EC7063", "#f9f9f9", "#45B39D"]
)

matriz_corr = df_transformado[cols_corr].corr()
plt.figure(figsize=(10, 6))
mask = np.triu(np.ones_like(matriz_corr, dtype=bool)) # Máscara para ver solo la mitad (evita redundancia)

sns.heatmap(matriz_corr,
            mask=mask,
            annot=True,
            fmt=".2f",
            cmap=cmap_personalizado,
            center=0,
            linewidths=.5,
            cbar_kws={"shrink": .8})

plt.title('Matriz de Correlación de Factores de Evasión', fontsize=16, fontweight='bold', pad=20)
plt.xticks(rotation=45, ha='right')
plt.show()
```

</p>
